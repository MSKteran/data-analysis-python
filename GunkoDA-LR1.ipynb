```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
P1 = [[1, 1], [1, 3], [3, 3], [3, 8], [6, 8], [6, 3], [3, 3], [8, 3], [8, 1]]
P2 = [[10, 1], [10, 8], [14, 8]]
```

```python
P1 = np.array(P1)
P2 = np.array(P2)

plt.plot(P1[:,0], P1[:,1])
plt.plot(P2[:,0], P2[:,1])
```

Отражение по оси Ox

```python
A = [[1,0], [0,-1]]
B1 = np.matmul(P1, A)
B2 = np.matmul(P2, A)

plt.plot(P1[:,0], P1[:,1])
plt.plot(P2[:,0], P2[:,1])
plt.plot(B1[:,0], B1[:,1])
plt.plot(B2[:,0], B2[:,1])
```

Поворот

```python
def turn(angle):
    Q = [[np.cos(angle), np.sin(angle)], [-np.sin(angle), np.cos(angle)]]
    return np.array(Q)
```

```python
# 120 градусов 
C1 = np.matmul(P1, turn(2 * np.pi / 3))
C2 = np.matmul(P2, turn(2 * np.pi / 3))
plt.plot(P1[:,0], P1[:,1])
plt.plot(P2[:,0], P2[:,1])
plt.plot(C1[:,0], C1[:,1])
plt.plot(C2[:,0], C2[:,1])
```

```python
# 240 градусов 
C1 = np.matmul(P1, turn(4 * np.pi / 3))
C2 = np.matmul(P2, turn(4 * np.pi / 3))
plt.plot(P1[:,0], P1[:,1])
plt.plot(P2[:,0], P2[:,1])
plt.plot(C1[:,0], C1[:,1])
plt.plot(C2[:,0], C2[:,1])
```

Увеличение рисунков

```python
# увеличение в 5 раз
plt.plot(P1[:,0], P1[:,1])
plt.plot(P2[:,0], P2[:,1])
plt.plot(5 * P1[:,0], 5 * P1[:,1])
plt.plot(5 * P2[:,0], 5 * P2[:,1])
```

Поворот увеличенного изображения на 60 градусов 5 раз

```python
# 1-й поворот
C1 = np.matmul(5 * P1, turn(np.pi / 3))
C2 = np.matmul(5 * P2, turn(np.pi / 3))

plt.plot(5 * P1[:,0], 5 * P1[:,1])
plt.plot(5 * P2[:,0], 5 * P2[:,1])
plt.plot(C1[:,0], C1[:,1])
plt.plot(C2[:,0], C2[:,1])
```

```python
# 2-й поворот
C1 = np.matmul(C1, turn(np.pi / 3))
C2 = np.matmul(C2, turn(np.pi / 3))

plt.plot(5 * P1[:,0], 5 * P1[:,1])
plt.plot(5 * P2[:,0], 5 * P2[:,1])
plt.plot(C1[:,0], C1[:,1])
plt.plot(C2[:,0], C2[:,1])
```

```python
# 3-й поворот
C1 = np.matmul(C1, turn(np.pi / 3))
C2 = np.matmul(C2, turn(np.pi / 3))

plt.plot(5 * P1[:,0], 5 * P1[:,1])
plt.plot(5 * P2[:,0], 5 * P2[:,1])
plt.plot(C1[:,0], C1[:,1])
plt.plot(C2[:,0], C2[:,1])
```

```python
# 4-й поворот
C1 = np.matmul(C1, turn(np.pi / 3))
C2 = np.matmul(C2, turn(np.pi / 3))

plt.plot(5 * P1[:,0], 5 * P1[:,1])
plt.plot(5 * P2[:,0], 5 * P2[:,1])
plt.plot(C1[:,0], C1[:,1])
plt.plot(C2[:,0], C2[:,1])
```

```python
# 5-й поворот
C1 = np.matmul(C1, turn(np.pi / 3))
C2 = np.matmul(C2, turn(np.pi / 3))

plt.plot(5 * P1[:,0], 5 * P1[:,1])
plt.plot(5 * P2[:,0], 5 * P2[:,1])
plt.plot(C1[:,0], C1[:,1])
plt.plot(C2[:,0], C2[:,1])
```

Умножение на произвольную матрицу

```python
D = [[-1, 1], [4, 0]]

M1 = np.matmul(P1, D)
M2 = np.matmul(P2, D)

plt.plot(P1[:,0], P1[:,1])
plt.plot(P2[:,0], P2[:,1])
plt.plot(M1[:,0], M1[:,1])
plt.plot(M2[:,0], M2[:,1])
```

Нахождение обратной матрицы для возвращение к исходному массиву

```python
I = np.linalg.inv(D)

P1 = np.matmul(M1, I)
P2 = np.matmul(M2, I)

plt.plot(M1[:,0], M1[:,1])
plt.plot(M2[:,0], M2[:,1])
plt.plot(P1[:,0], P1[:,1])
plt.plot(P2[:,0], P2[:,1])
```
