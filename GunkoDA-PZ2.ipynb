```python
import numpy as np

import matplotlib.pyplot as plt
```

1 задание — инициалы Д Г с помощью графиков параболы и эллипса

```python
# Параметры для буквы "Д" (Используем график параболы и эллипса)

# Левая ножка параболы
X = np.linspace(-0.5, 0)
Y = 0.25 + 3 * (X + 0.5)**2

# Верхняя дуга эллипса
T = np.linspace(0, np.pi)
X1 = 1.5 + np.cos(T)
Y1 = 1 + 1 * np.sin(T)

X = np.append(X, X1)
Y = np.append(Y, Y1)

# Правая ножка
X2 = np.linspace(3, 3.5)
Y2 = 0.25 + 3 * (X2 - 3.5) ** 2

X = np.append(X, X2)
Y = np.append(Y, Y2)

# Параметры для буквы "Г"

# Вертикальная ножка параболы
X3 = np.linspace(4, 4.5)
Y3 = 0.25 + 7 * (X3 - 4) ** 2

X = np.append(X, X3)
Y = np.append(Y, Y3)

# Горизонтальная ножка
X4 = np.linspace(4.5, 7)
Y4 = 2 * np.ones_like(X4)  # Горизонтальная линия на уровне Y=1.5

X = np.append(X, X4)
Y = np.append(Y, Y4)

plt.plot(X,Y)
plt.grid()
plt.show()
```

```python
# Функция градиентного спуска
def grad_opt(f, grad_f, x, y, step, iter):
    opt = np.zeros((iter,2))
    for i in range(iter):
        opt[i, :] = x, y
        L = (grad_f(x,y)[0] ** 2 + grad_f(x,y)[1] ** 2) ** 0.5
        if L > 0:
            x += -grad_f(x,y)[0] / L * step
            y += -grad_f(x,y)[1] / L * step
    return opt
```

```python
# f = 3x^2 + xy + y^2 + x + y
def f(x, y):
    return 3 * x ** 2 + x * y + y ** 2 + x + y

# компоненты вектора градиента
def grad_f(x, y):
    return 6 * x + y + 1, x + 2 * y + 1
```

```python
res = grad_opt(f, grad_f, 2, -1, 0.001, 6000)
plt.scatter(res[:,0], res[:,1])

res1 = grad_opt(f, grad_f, -2, 1, 0.001, 6000)
plt.scatter(res1[:,0], res1[:,1])

res2 = grad_opt(f, grad_f, -2, -1, 0.001, 6000)
plt.scatter(res2[:,0], res2[:,1])

plt.grid()
plt.show()
```

```python
print(f"{res[-1,0]:.3f}", f"{res[-1,1]:.3f}")
print(f"{res1[-1,0]:.3f}", f"{res1[-1,1]:.3f}")
print(f"{res2[-1,0]:.3f}", f"{res2[-1,1]:.3f}")
```
