```python
# Вариант 4. Построить диаграмму зависимости цены на 5-комнатную квартиру от этажа, файл 5k.xlsx
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из файла Excel
df = pd.read_excel("5k.xlsx")

# Удаление строк с NaN в столбцах "Этаж" и "Цена, т.р."
df.dropna(subset=['Этаж', 'Цена, т.р.'], inplace=True)

# Извлечение этажа (числа до знака "/")
df['Этаж'] = df['Этаж'].str.split('/').str[0].astype(int)

# Сохранение только необходимых столбцов
df = df[['Этаж', 'Цена, т.р.']]
```

```python
# Вывод таблицы
print(df)
```

```python
# Диаграмма рессеяния
plt.figure(figsize=(10, 6))
plt.scatter(df['Этаж'], df['Цена, т.р.'], s=50)
plt.xlabel('Этаж', fontsize=14)
plt.ylabel('Цена, т.р.', fontsize=14)
plt.title('Зависимость цены от этажа', fontsize=16)
plt.show()
```

```python
# Столбчатая диаграмма
plt.figure(figsize=(10, 6))
plt.bar(df['Этаж'], df['Цена, т.р.'])
plt.xlabel('Этаж', fontsize=14)
plt.ylabel('Цена, т.р.', fontsize=14)
plt.title('Зависимость цены от этажа (bar)', fontsize=16)
plt.show()
```

```python
# Гистограмма
plt.figure(figsize=(10, 6))
plt.hist(df['Цена, т.р.'], bins=10)
plt.xlabel('Цена, т.р.', fontsize=14)
plt.ylabel('Частота', fontsize=14)
plt.title('Гистограмма распределения цен', fontsize=16)
plt.show()
```
